```python
import pandas as pd
import numpy as np
import os
import sys
import pickle

import prody
import torch
import dgl
from dgl.dataloading import GraphDataLoader
import prody as pr
from tqdm.auto import tqdm

import torch
from torch.optim import Adam
from torch.optim.lr_scheduler import StepLR, ReduceLROnPlateau
import torch.nn.functional as F

prody.confProDy(verbosity='none')
sys.path.append('/net/galaxy/home/koes/vratins/graphpocket/gnn')
from graphpocket import GraphPocket
from model import ReceptorEncoderGVP, con_loss, get_batch_idx

from sklearn.model_selection import KFold, GroupShuffleSplit
from sklearn.cluster import DBSCAN, KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.utils import shuffle
from sklearn.cluster import AgglomerativeClustering

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
class GraphTupleInference(dgl.data.DGLDataset):

    def __init__(self, name, pocket_list, k, algorithm):
        
        self.pocket_list = pocket_list
        self.k = k
        self.algorithm = algorithm
        self.pocket_index_map = []

        super().__init__(name=name)

    def __getitem__(self, index):        

        return self.pocket_index_map[index]

    def __len__(self):

        return len(self.pocket_index_map)
    
    def process(self):

        pocket_to_graph = GraphPocket(k = self.k, algorithm = self.algorithm)

        for i, pocket in tqdm(enumerate(self.pocket_list), desc=f"Reading {self.name} Pockets"):
            graph = pocket_to_graph(pocket_path=pocket)
            pocket_dict = {}
            pocket_dict['ID'] = pocket.split('/')[-1].split('.')[0]
            pocket_dict['path'] = pocket
            pocket_dict['graph'] = graph
            self.pocket_index_map.append(pocket_dict)

def create_dataset(pocket_list, k, algorithm, seed=42):

    print("reading pockets...")
    dataset = GraphTupleInference(name='test', pocket_list=pocket_list, k=k, algorithm=algorithm) 

    return dataset

def collate_fn(samples):
    graphs = map(list, zip(*samples))
    
    batched_graph = dgl.batch(graphs)
    
    return batched_graph

def get_dataloader(dataset, batch_size, num_workers, **kwargs):

    dataloader = GraphDataLoader(dataset, batch_size=batch_size, drop_last=False, num_workers=num_workers, **kwargs)

    return dataloader
```

## GraphPocket for KER

```python
kinome_dir_7 = '../../kinome/kinome_pockets_7A/'
estrogen_dir_7 = '../../kinome/estrogen_pockets_7A/'
kinome_list_7 = os.listdir(kinome_dir_7)
estrogen_list_7 = os.listdir(estrogen_dir_7)
kinome_list_7 = [os.path.join(kinome_dir_7, kinome) for kinome in kinome_list_7]
estrogen_list_7 = [os.path.join(estrogen_dir_7, est) for est in estrogen_list_7]
total_pockets_7 = kinome_list_7 + estrogen_list_7

test_dataset = create_dataset(total_pockets_7, 6, 'kd-tree')
test_dataloader = get_dataloader(test_dataset, 64, 2)

print(total_pockets_7)
```

```python
#load in model

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = ReceptorEncoderGVP(
    in_scalar_size=4,    
    out_scalar_size=48,
    edge_feat_size = 1,
    n_message_gvps = 1,
    n_update_gvps = 1,
    vector_size=16,      
    n_convs=2,           
    dropout=0.2,
    message_norm = 0
)

checkpoint = torch.load('../../graphpocket/results/model_final_9.pth.tar') 
model.load_state_dict(checkpoint['model_state_dict'])
model.to(device)

model.eval()

pocket_map = {'path':[],'descriptor':[]}

progress_bar = tqdm(enumerate(test_dataloader), total=len(test_dataloader))

with torch.no_grad():
    for batch_idx, pockmap in progress_bar:

        graph = pockmap['graph']
        graph = graph.to(device)

        batch_idx = get_batch_idx(graph).to(device)

        output = model(graph, batch_idx)

        pocket_map['path'].extend(pockmap['path'])
        pocket_map['descriptor'].extend(output.cpu().numpy())
```

```python
descriptors = pocket_map['descriptor']

labels = []

for path in pocket_map['path']:
    if 'kinome_pockets' in path:
        labels.append('Kinase')
    else:
        labels.append('Estrogen Receptor')

print(len(labels))
```

```python
scaler = StandardScaler()
desc_scaled = scaler.fit_transform(descriptors)

pca = PCA(n_components=5)  
components = pca.fit_transform(desc_scaled)

print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=components[:, 0], y=components[:, 1], hue=labels, palette="tab10", s=50, legend=True)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Descriptors - GraphPocket 7A')
plt.legend()
plt.savefig('graph7a.png')
plt.show()
```
